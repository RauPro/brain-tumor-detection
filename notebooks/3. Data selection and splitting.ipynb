```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import random
```

```python
base_dir = "../data/processed"
train_dir = os.path.join(base_dir, 'train')
validation_dir = os.path.join(base_dir, 'validation')
```

```python
fig =plt.figure(figsize=(28,14))
columns=5
rows=1
healthy_trn_samples = (os.path.join(train_dir, 'no'))
infected_trn_samples = (os.path.join(train_dir, 'yes'))

infected_trn_fpaths = [os.path.join(infected_trn_samples, fpath) for
                       fpath in os.listdir(infected_trn_samples)]

healthy_trn_fpaths = [os.path.join(healthy_trn_samples, fpath) for
                      fpath in os.listdir(healthy_trn_samples)]
random.shuffle(healthy_trn_fpaths)
random.shuffle(infected_trn_fpaths)
#shuffle lists


for i in range(1, columns*rows +1):
    fig.add_subplot(rows, columns, i)
    plt.imshow(mpimg.imread(infected_trn_fpaths[i-1]))
    plt.axis('off')
    plt.title('Tumorous', fontsize=32)
plt.show()


fig =plt.figure(figsize=(28,14))
columns=5
rows=1
for i in range(1, columns*rows +1):
    fig.add_subplot(rows, columns, i)
    plt.imshow(mpimg.imread(healthy_trn_fpaths[i-1]))
    plt.axis('off')
    plt.title('Non - Tumorous', fontsize=32)
plt.show()
```
